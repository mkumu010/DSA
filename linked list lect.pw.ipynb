```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

a=Node(5)
b=Node(4)
c=Node(7)

a.next=b
b.next=c

head=a
print(head.data)
print(head.next.data)
print(head.next.next.data)
        


```

```python
#traverse the linked list
def printlinkedlist(head):
    curr=head
    while curr!=None:
        print(curr.data,end=" ")
        curr=curr.next
printlinkedlist(head)
```

```python
#insertion at the beginning of the linked list

newNode=Node(2)

newNode.next=head
head=newNode

printlinkedlist(head)

```

```python
#insertion at the end of the linked list

newNode=Node(8)
curr=head
while curr.next!=None:
    curr=curr.next
curr.next=newNode
printlinkedlist(head)


```

```python
#insertion at the middle(kth) of the linked list

k=2
newNode=Node(6)
curr=head
for i  in range(k-1):
    curr=curr.next
newNode.next=curr.next
curr.next=newNode
printlinkedlist(head)
```

```python
#delete the first node

head=head.next
printlinkedlist(head)
```

```python
#delete the last node

curr=head
while curr.next.next!=None:
    curr=curr.next
curr.next=None
printlinkedlist(head)

    
```

```python
#delete the middle(kth) node

curr=head
for i in range(k-1):
    curr=curr.next
curr.next=curr.next.next
printlinkedlist(head)
```
